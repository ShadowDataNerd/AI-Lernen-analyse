```python
import pandas as pd
from bokeh.plotting import output_notebook, figure, show
from bokeh.io import show, output_file, output_notebook, reset_output
from bokeh.models import ColumnDataSource, FactorRange
from bokeh.transform import factor_cmap
from bokeh.models import HoverTool
reset_output()
output_notebook()
```

```python
patentdata = pd.read_csv("W:\Desktop\Business Intelligence\\AnzahlPatenteProJahrMitLernkategorieBalkendiagramm.csv",
                       sep = ";", names = ["AnzahlPatente", "Jahr", "Lernkategorie"])
patentdata = patentdata.astype({"Jahr": str})
patentdata.head()
```

```python
patentdata['x'] = list(zip(patentdata.Jahr,patentdata.Lernkategorie))
```

```python
patent_source = patentdata
```

```python
patentdata["Jahr"].unique()
```

```python
hover = HoverTool(tooltips=[("Anzahl", "@AnzahlPatente")])
tools = [hover, "pan","wheel_zoom","box_zoom","reset", "save", "box_select"]
```

```python
source = ColumnDataSource(patentdata)
r = patentdata["x"]
palette = ["red", "blue", "green", "yellow", "orange", "brown", "black"]
p = figure(plot_width=2000, plot_height = 500, title = "Anzahl Patente und Orga", x_range = FactorRange(*r), tools = tools)
p.vbar(x = "x", top = 'AnzahlPatente', width = 1.5,  source = source,        
        fill_color = factor_cmap('x', palette = palette,
                                 factors = ["Anzahl Organisation","Patente insgesamt","Active Learning", "Reinforcement Learning",
                                            "BSemi-Supervised Learning", "Supervised Learning", "Unsupervised Learning"],
                                 start = 1, end = 2),
       #legend = [value(x) for x in patentdata.columns]
      )
#p.legend.location = "top"
#p.legend.click_policy = "hide"
p.xaxis.major_label_orientation = 1.0
show(p)
```
